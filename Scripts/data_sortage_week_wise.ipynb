```python
import os
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
import pandas as pd

# Define the top-level directory
top_directory = "/media/ashraf/Seagate/PROCESSING/NEPAL"

# List of station codes to include
station_codes = {"BESI", "DMAU", "KKN4", "KIRT", "NAST", "SNDL", "CHLM", "SIM4", "HETA", "SLBL", 
                 "LMJG", "CHWN", "GHER", "SYBC", "MKLU", "JIR2", "XYAK", "BNDP"}

# Function to check if a date is within a 7-day interval
def is_within_week_interval(date, start):
    return start <= date < start + timedelta(days=7)

# Function to process data for a given year
def process_year(year):
    # Dictionary to store the results for each station
    station_results = {code: [] for code in station_codes}

    # Define the relevant file extensions for the year
    file_extensions = [f'.{str(year)[2:]}o', f'.{str(year)[2:]}d', f'.{str(year)[2:]}o.Z', f'.{str(year)[2:]}d.Z']

    # Walk through the directory tree
    for root, dirs, files in os.walk(top_directory):
        # Process only directories that contain files with the specified extensions
        relevant_files = [f for f in files if any(f.endswith(ext) for ext in file_extensions)]
        
        if relevant_files:
            # Define the start date for the first interval
            start_date = datetime(year, 1, 1)

            # Loop through the year in 7-day intervals
            current_start_date = start_date
            while current_start_date.year == year:
                # Corresponding interval six months later
                corresponding_start_date = current_start_date + relativedelta(months=+6)

                # Process files in the current interval
                for f in relevant_files:
                    # Extract the station code from the filename
                    station_code = f[:4].upper()
                    if station_code not in station_codes:
                        continue

                    # Extract the DOY from the filename and convert it to a datetime object
                    try:
                        doy = int(f[4:7])
                        date = datetime(year, 1, 1) + timedelta(days=doy - 1)
                    except ValueError:
                        continue  # Skip files with invalid DOY

                    # Count files in the current 7-day interval
                    if is_within_week_interval(date, current_start_date):
                        station_results[station_code].append([
                            current_start_date.strftime('%Y-%m-%d'),
                            1,
                            corresponding_start_date.strftime('%Y-%m-%d'),
                            0
                        ])

                    # Count files in the corresponding 7-day interval six months later
                    if is_within_week_interval(date, corresponding_start_date):
                        station_results[station_code].append([
                            current_start_date.strftime('%Y-%m-%d'),
                            0,
                            corresponding_start_date.strftime('%Y-%m-%d'),
                            1
                        ])
                
                current_start_date += timedelta(days=7)

    return station_results

# Create a Pandas Excel writer using XlsxWriter as the engine
output_file = "Weekly_data_comparison_highest_2011_2016.xlsx"
with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
    # Dictionary to store the highest file counts for comparison
    comparison_data = []

    # Process each year from 2011 to 2016
    for year in range(2011, 2017):
        station_results = process_year(year)

        # Process each station's results
        for station_code in station_codes:
            intervals = station_results[station_code]
            # Aggregate the counts for each interval
            aggregated_intervals = {}
            for interval in intervals:
                key = (interval[0], interval[2])
                if key not in aggregated_intervals:
                    aggregated_intervals[key] = [interval[0], 0, interval[2], 0]
                aggregated_intervals[key][1] += interval[1]
                aggregated_intervals[key][3] += interval[3]

            # Convert the dictionary to a sorted list
            sorted_intervals = sorted(aggregated_intervals.values(), key=lambda x: (x[0], x[2]))

            if sorted_intervals:
                # Find the interval with the maximum number of files in January
                max_jan_interval = max(sorted_intervals, key=lambda x: x[1])

                # Add to comparison data
                comparison_data.append([
                    station_code,
                    max_jan_interval[0],
                    max_jan_interval[1],
                    max_jan_interval[2],
                    max_jan_interval[3]
                ])

    # Create a DataFrame and write the comparison data to the Excel file
    comparison_df = pd.DataFrame(comparison_data, columns=["Station Code", "Week Starting Date (Jan)", "Number of Files (Jan)", "Week Starting Date (Jun)", "Number of Files (Jun)"])
    comparison_df.to_excel(writer, sheet_name="Comparison", index=False)

print(f"Results saved to {output_file}")
```

```python
import os
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
import pandas as pd

# Define the top-level directory
top_directory = "/media/ashraf/Seagate/PROCESSING/NEPAL"

# List of station codes to include
station_codes = {"BESI", "DMAU", "KKN4", "KIRT", "NAST", "SNDL", "CHLM", "SIM4", "HETA", "SLBL", 
                 "LMJG", "CHWN", "GHER", "SYBC", "MKLU", "JIR2", "XYAK", "BNDP"}

# Function to check if a date is within a 7-day interval
def is_within_week_interval(date, start):
    return start <= date < start + timedelta(days=7)

# Function to process data for a given year
def process_year(year):
    # Dictionary to store the results for each station
    station_results = {code: [] for code in station_codes}

    # Define the relevant file extensions for the year
    file_extensions = [f'.{str(year)[2:]}o', f'.{str(year)[2:]}d', f'.{str(year)[2:]}o.Z', f'.{str(year)[2:]}d.Z']

    # Walk through the directory tree
    for root, dirs, files in os.walk(top_directory):
        # Process only directories that contain files with the specified extensions
        relevant_files = [f for f in files if any(f.endswith(ext) for ext in file_extensions)]
        
        if relevant_files:
            # Define the start date for the first interval
            start_date = datetime(year, 1, 1)

            # Loop through the year in 7-day intervals
            current_start_date = start_date
            while current_start_date.year == year:
                # Corresponding interval six months later
                corresponding_start_date = current_start_date + relativedelta(months=+6)

                # Process files in the current interval
                for f in relevant_files:
                    # Extract the station code from the filename
                    station_code = f[:4].upper()
                    if station_code not in station_codes:
                        continue

                    # Extract the DOY from the filename and convert it to a datetime object
                    try:
                        doy = int(f[4:7])
                        date = datetime(year, 1, 1) + timedelta(days=doy - 1)
                    except ValueError:
                        continue  # Skip files with invalid DOY

                    # Count files in the current 7-day interval
                    if is_within_week_interval(date, current_start_date):
                        station_results[station_code].append([
                            current_start_date.strftime('%Y-%m-%d'),
                            1,
                            corresponding_start_date.strftime('%Y-%m-%d'),
                            0
                        ])

                    # Count files in the corresponding 7-day interval six months later
                    if is_within_week_interval(date, corresponding_start_date):
                        station_results[station_code].append([
                            current_start_date.strftime('%Y-%m-%d'),
                            0,
                            corresponding_start_date.strftime('%Y-%m-%d'),
                            1
                        ])
                
                current_start_date += timedelta(days=7)

    return station_results

# Function to find the common highest interval across all stations for a given year
def find_common_highest_interval(station_results):
    # Dictionary to store interval counts
    interval_counts = {}

    # Calculate total counts for each interval across all stations
    for station_code, intervals in station_results.items():
        for interval in intervals:
            key = (interval[0], interval[2])  # Using Jan and Jun start dates as key
            if key not in interval_counts:
                interval_counts[key] = [0, 0]  # Initialize Jan and Jun counts
            interval_counts[key][0] += interval[1]  # Add Jan count
            interval_counts[key][1] += interval[3]  # Add Jun count

    # Check if interval_counts is empty
    if not interval_counts:
        raise ValueError("No intervals found with file counts")

    # Find the interval with the maximum number of files in January
    max_jan_interval = max(interval_counts.items(), key=lambda x: x[1][0])[0]

    return max_jan_interval

# Create a Pandas Excel writer using XlsxWriter as the engine
output_file = "Common_Highest_Files_Intervals.xlsx"
with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
    # Process each year from 2011 to 2016
    for year in range(2011, 2017):
        try:
            station_results = process_year(year)

            # Find the common highest interval for this year
            common_interval = find_common_highest_interval(station_results)

            # Prepare data for each station using the common interval
            data = []
            for station_code in station_codes:
                interval_data = station_results[station_code]
                # Find the specific interval for this station that matches the common interval
                for interval in interval_data:
                    if (interval[0], interval[2]) == common_interval:
                        data.append([
                            station_code,
                            interval[0],  # Week starting date (Jan)
                            interval[1],  # Number of files (Jan)
                            interval[2],  # Week starting date (Jun)
                            interval[3]   # Number of files (Jun)
                        ])
                        break  # Once found, break out of the loop

            # Create a DataFrame and write the data to the Excel file for this year
            df = pd.DataFrame(data, columns=["Station Code", "Week Starting Date (Jan)", "Number of Files (Jan)", "Week Starting Date (Jun)", "Number of Files (Jun)"])
            df.to_excel(writer, sheet_name=str(year), index=False)

            print(f"Processed year {year} successfully")
        except Exception as e:
            print(f"Error processing year {year}: {str(e)}")
            continue  # Skip to the next year if there's an error

print(f"Results saved to {output_file}")
```

```python
import os
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
import pandas as pd

# Define the top-level directory
top_directory = "/media/ashraf/Seagate/PROCESSING/NEPAL"

# List of station codes to include
station_codes = {"BESI", "DMAU", "KKN4", "KIRT", "NAST", "SNDL", "CHLM", "SIM4", "HETA", "SLBL", 
                 "LMJG", "CHWN", "GHER", "SYBC", "MKLU", "JIR2", "XYAK", "BNDP"}

# Function to check if a date is within a 7-day interval
def is_within_week_interval(date, start):
    return start <= date < start + timedelta(days=7)

# Dictionary to store the results for each station
station_results = {code: [] for code in station_codes}

# Walk through the directory tree
for root, dirs, files in os.walk(top_directory):
    # Process only directories that contain files with the specified extensions
    relevant_files = [f for f in files if f.endswith('.15o') or f.endswith('.15d') or f.endswith('.15o.Z') or f.endswith('.15d.Z')]
    
    if relevant_files:
        # Define the start date for the first interval
        start_date = datetime(2015, 1, 1)

        # Loop through the year in 7-day intervals
        current_start_date = start_date
        while current_start_date.year == 2015:
            # Corresponding interval six months later
            corresponding_start_date = current_start_date + relativedelta(months=+6)

            # Process files in the current interval
            for f in relevant_files:
                # Extract the station code from the filename
                station_code = f[:4].upper()
                if station_code not in station_codes:
                    continue

                # Extract the DOY from the filename and convert it to a datetime object
                try:
                    doy = int(f[4:7])
                    date = datetime(2015, 1, 1) + timedelta(days=doy - 1)
                except ValueError:
                    continue  # Skip files with invalid DOY

                # Count files in the current 7-day interval
                if is_within_week_interval(date, current_start_date):
                    station_results[station_code].append([
                        current_start_date.strftime('%Y-%m-%d'),
                        1,
                        corresponding_start_date.strftime('%Y-%m-%d'),
                        0
                    ])

                # Count files in the corresponding 7-day interval six months later
                if is_within_week_interval(date, corresponding_start_date):
                    station_results[station_code].append([
                        current_start_date.strftime('%Y-%m-%d'),
                        0,
                        corresponding_start_date.strftime('%Y-%m-%d'),
                        1
                    ])
            
            current_start_date += timedelta(days=7)

# Function to find the station with the highest count for each week
def find_highest_counts(station_results, week_start_date, week_end_date):
    week_counts = {}
    for station, intervals in station_results.items():
        for interval in intervals:
            if interval[0] == week_start_date:
                if station not in week_counts:
                    week_counts[station] = interval[1]
                else:
                    week_counts[station] += interval[1]
            if interval[2] == week_end_date:
                if station not in week_counts:
                    week_counts[station] = interval[3]
                else:
                    week_counts[station] += interval[3]
    if week_counts:
        return max(week_counts, key=week_counts.get), week_counts
    else:
        return None, week_counts

# Create a Pandas Excel writer using XlsxWriter as the engine
output_file = "Weekly_data_ashraf.xlsx"
with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
    # Process each station's results
    for station_code in station_codes:
        intervals = station_results[station_code]
        # Aggregate the counts for each interval
        aggregated_intervals = {}
        for interval in intervals:
            key = (interval[0], interval[2])
            if key not in aggregated_intervals:
                aggregated_intervals[key] = [interval[0], 0, interval[2], 0]
            aggregated_intervals[key][1] += interval[1]
            aggregated_intervals[key][3] += interval[3]

        # Convert the dictionary to a sorted list
        sorted_intervals = sorted(aggregated_intervals.values(), key=lambda x: (x[0], x[2]))

        # Create a DataFrame and write to the Excel file
        df = pd.DataFrame(sorted_intervals, columns=["Week Starting Date (Jan-Jun)", "Number of Files (Jan)", "Week Starting Date (Jul-Dec)", "Number of Files (Jun)"])
        df.to_excel(writer, sheet_name=station_code, index=False)

        # Apply conditional formatting to highlight the maximum values
        workbook  = writer.book
        worksheet = writer.sheets[station_code]
        format_highlight = workbook.add_format({'bg_color': '#FFC7CE', 'font_color': '#9C0006'})

        # Find the max values in the 'Number of Files (Jan)' and 'Number of Files (Jun)' columns
        max_jan = df["Number of Files (Jan)"].max()
        max_jun = df["Number of Files (Jun)"].max()

        # Apply conditional formatting
        worksheet.conditional_format('B2:B{}'.format(len(df) + 1), {'type': 'cell', 'criteria': '==', 'value': max_jan, 'format': format_highlight})
        worksheet.conditional_format('D2:D{}'.format(len(df) + 1), {'type': 'cell', 'criteria': '==', 'value': max_jun, 'format': format_highlight})

# Print the comparison results
start_date = datetime(2015, 1, 1)
current_start_date = start_date
while current_start_date.year == 2015:
    corresponding_start_date = current_start_date + relativedelta(months=+6)
    highest_count_station, week_counts = find_highest_counts(station_results, current_start_date.strftime('%Y-%m-%d'), corresponding_start_date.strftime('%Y-%m-%d'))
    if highest_count_station:
        print(f"Week starting {current_start_date.strftime('%Y-%m-%d')} (Jan) and {corresponding_start_date.strftime('%Y-%m-%d')} (Jun): Highest count station: {highest_count_station} with counts {week_counts}")
    else:
        print(f"Week starting {current_start_date.strftime('%Y-%m-%d')} (Jan) and {corresponding_start_date.strftime('%Y-%m-%d')} (Jun): No data found")
    current_start_date += timedelta(days=7)

print(f"Results saved to {output_file}")
```
